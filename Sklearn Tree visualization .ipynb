```python
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree, export_text
```

```python
#using a diabetes for example
import pandas as pd
df = pd.read_csv('diabetes.csv')
# setting the X , y 
X=df.drop(columns=['Outcome'])
y=df['Outcome']
```

```python
features=list(df.columns[0:-1])
class_name=["non_diabetic", "diabetic"]
```

```python
dt = DecisionTreeClassifier(max_depth=3, random_state=3)
```

```python
dt.fit(X, y)
```

```python
plt.figure(figsize=(15,15))
plot_tree(dt, feature_names=features, filled =True, class_names=["0","1"]);
```

```python
print(export_text(dt, feature_names=features, show_weights=True))
```
