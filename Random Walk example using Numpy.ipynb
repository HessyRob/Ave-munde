```python
#import libs
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_0 = np.array([[0],[0]]) # starting point origin 
```

```python
delta_x = np.random.normal(0,1,(2,1000)) 
#displacements from random Normal distribution , generate array shape 2 x 1000 , using numbers between 0 and 1 
```

```python
X = np.concatenate((X_0,np.cumsum(delta_x, axis=1)), axis=1) # X values with displacements 
```

```python
X.shape
# this is original point + each point and random displacement 
```

```python
plt.plot(X[0], X[1], "ro-" , markeredgecolor ="r", markersize=5, linewidth=0.5) #plot X coordinates from each column
```
